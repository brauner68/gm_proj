# Config + Train

```python
from src.trainer import DiffusionTrainer
from src.config import get_config
from pathlib import Path

home = Path.home() / 'gm_proj'
data_path = home / 'data/nsynth_data/nsynth-valid'
out_dir = home / 'results/runai_results/01'

# 1. Configuration
config = get_config()
config['output_dir'] = out_dir
config['data_path'] = data_path
config['lr'] = 1e-5
config['max_samples'] = None
config['selected_families'] = ['string', 'mallet']
config['batch_size'] = 32
config['epochs'] = 15
config['save_interval'] = 5
config['cfg_prob'] = 0.1
config['T_target'] = 80
config['conditioning'] = 'concat'

# 2. Run
print("üöÄ Starting Training...")
trainer = DiffusionTrainer(config)
trainer.train()
trainer.plot_loss_curve()
```

```python
import IPython.display as ipd
from src.generator import DiffusionGenerator


ckpt_path = out_dir / 'checkpoint_epoch_19.pt'
gen = DiffusionGenerator(config, ckpt_path)

audio_results = gen.generate(samples_per_class=1) 

# 3. Listen!
print("\nüéß --- Listening Room ---")
for instrument, waveforms in audio_results.items():
    print(f"\nüé∏ Instrument: {instrument}")
    for i, wav in enumerate(waveforms):
        # wav is [1, T], ipd expects [T] numpy array
        audio_data = wav.squeeze().numpy()
        
        # Display Audio Widget
        ipd.display(ipd.Audio(audio_data, rate=22050))
```
